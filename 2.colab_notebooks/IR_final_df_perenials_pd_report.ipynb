```
import pandas as pd
from pandas_profiling import ProfileReport
```

```
data_raw = pd.read_csv('perenials_dataset1.csv', encoding='utf8', header=0, delimiter=';')
```

```
data_raw.head()
```

```

profile = ProfileReport(data_raw, title = 'IR_final_perenails_raw_pandas_profiling_teport', html={'style':{'full_width':False}})
profile
```

```
profile.to_notebook_iframe()
```

```
profile.to_file(output_file="IR_final_perenials_report.html")
```
